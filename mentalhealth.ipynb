```python

help_dict = {
    "sad": True,
    "depressed": True,
    "alone": True
}


user_post = input("How are you feeling today? ").lower()


if any(word in user_post for word in help_dict):
    print("It sounds like you're going through a tough time.")
    print("Please call the student support line at 1-800-HELP")

else:
    print("Let's do a quick mood screening.")


    mood = int(input("Rate your mood today (1-10): "))
    energy = int(input("Rate your energy level (1-10): "))
    sleep = int(input("How well did you sleep last night (1-10): "))


    total_score = mood + energy + sleep
    average_score = total_score / 3
    wellness_percentage = (total_score / 30) * 100  # max possible score = 30


    print("\nYour total score is:", total_score)
    print("Your average score is:", round(average_score, 2))
    print("Your wellness percentage is:", round(wellness_percentage, 1), "%")


    if average_score <= 3:
        print("You may be having a very difficult day.")
        print("Consider reaching out to a counselor or support service.")
    elif average_score <= 6:
        print("It seems like you might be having a challenging day.")
        print("Try doing something relaxing or talking with someone you trust.")
    else:
        print("You seem to be doing fairly well today.")
        print("Continue to main menu.")

print("\nThank you for completing the screening. Continue to main menu.")

def main_menu():
  while True:
    print("\n=== MAIN MENU ===")
    print("1) Mental health screener (diagnosis)")
    print("2) Local group finder")
    print("3) Therapist and counselor finder")
    print("4) Psychiatrist finder")
    print("5) Quit")

    choice = input("Choose 1-6: ").strip()

    if choice == "1":
      def diagnose_mental_health():
        diagnoses = {"generalized Anxiety Disorer": ["restlessness", "racing thoughts", "worrying", "insomniea"],
                     "depressive disorder": ["low mood", "fatigue", "loss of instrest", "appitie changes"],
                     "ADHD": ["easily distracted", "impulsive", "forgetfulness", "fidgeting"]}

        print("--- Mental Health Screener ---")
        print("please enter your symptoms seperated by commas:")
        user_symptoms = [s.strip() for s in input(">").lower().split(",")]
        results = {}
        for condition, symptoms in diagnoses.items():
          matches = set(user_symptoms).intersection(set(symptoms))
          match_count = len(matches)
          if match_count > 0:
            results[condition] = match_count

        print("\n--- potentiL indications ---")

        if not results:
          print("No matches found based on the symptoms provided")
        else:
          sorted_results = sorted(results.items(), key=lambda x: x[1], reverse=True)
          for condition, count in sorted_results:
            print(f"- {condition}: {count} symptoms(s) matched")
      diagnose_mental_health()
    elif choice == "2":
      print("We understand you're going through a tough time and would like to meet with a local group.")
      print("Let's us help you get connected with a group near you.")
      print("Please select one of the following choices as it's most convenient for you:")
      # Assuming local_group() function exists elsewhere
      # local_group()
    elif choice == "3":
      print("We understand you're going through a tough time and would like to meet with a therapist/counselor.")
      print("Let's us help you get connected with one near you.")
      print("Please select one of the following choices as it's most convenient for you:")
      # Assuming therapist_counseling() function exists elsewhere
      # therapist_counseling()
    elif choice == "4":
      print("We understand you're going through a tough time and would like to meet with a psychiatrist.")
      print("Let's us help you get connected with one near you.")
      print("Please select one of the following choices as it's most convenient for you:")
      # Assuming pychiatrist_finder() function exists elsewhere
      # pychiatrist_finder()
    elif choice == "5":
      print("Goodbye!")
      break
    else:
      print("Invalid choice. Please choose between 1 and 5.")
main_menu()
```
